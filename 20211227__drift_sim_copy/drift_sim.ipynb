```python

# generating the simulation data for the bottom left plot


if [False,True][1]: # NOTE: only set to True if you'd like to re-calculate the simulation data
    t_i = datetime.datetime.now()
    e_field_sim_list = ["hemisphere_1000V.txt"]
    e_field_sim_list = ["hemisphere_1000V.txt", "flat_sphere_1000V.txt", "sphere_1000V.txt", "small_zylinder_1000V.txt", "medium_zylinder_1000V.txt", "large_zylinder_1000V.txt"]
    sensor_pos_list = [[0.005, 0.0888], [0.005, 0.0888], [0.005, 0.0888]]
    sensor_pos_list = [get_std_sensor_pos_from_comsol_sim_array(abspath_this_study+e_field_sim) for e_field_sim in e_field_sim_list]
    #print(f"sensor_pos_list = {sensor_pos_list}")
    u_coll_v_list = [100, 300, 1000]
    neutral_halflife_list = [1, 10**6]
    n_sim = 5000 # number of particle simulations per e_field_sim and u_coll_v
    simulation_outcome_tuple_list = []
    simulation_outcome_dtype = np.dtype([
        ("e_field_sim", np.unicode_, 16),
        ("u_coll_v", np.float64),
        ("eff_po218", np.float64),
        ("eff_po214", np.float64),
        ("drift_time_mean", np.float64),
        ("drift_time_lower", np.float64),
        ("drift_time_upper", np.float64),
        ("neutral_halflife", np.float64),
    ])


    # looping over all field simulations and voltages
    for i, e_field_sim in enumerate(e_field_sim_list):
        for j, u_coll_v in enumerate(u_coll_v_list):
            for k, neutral_halflife in enumerate(neutral_halflife_list):


                # collection efficiency simulation for every collection voltage value
                particle_sim_tuple_list = []
                for j in range(n_sim):
                    p = drift_sim.simulate_1p(
                        file = abspath_this_study +e_field_sim,
                        modfac = u_coll_v/1000, # linearly scale the electrical field strength to only require one input file
                        sensor_pos = sensor_pos_list[i],
                        neutral_halflife = neutral_halflife,
                        neutral = True,
                        diff = True)
                    particle_sim_tuple_list.append(( p.flag, p.time, p.nucleus ))
                    if j%1000==0 or j==100: print(f"{j}/{n_sim}")
                    #print(f"sim res #{j} @{u_coll}V: flag='{p.flag}', nucleus='{p.nucleus}', drift_time={p.time}")


                # evaluating the outcome of a collection voltage simulation
                particle_sim_array = np.array(particle_sim_tuple_list, np.dtype([("flag", np.float16),("drift_time", np.float64),("nucleus", np.unicode_, 16)]))
                print(f"e_field_sim='{e_field_sim}', U_coll={u_coll_v}V, sensor_pos={sensor_pos_list[i]}, neut={neutral_halflife}:")
                # flags
                print(f"\tflag: n0={len(particle_sim_array[(particle_sim_array['flag']==0)])} (in flight), n1={len(particle_sim_array[(particle_sim_array['flag']==1)])} (arrived), n2={len(particle_sim_array[(particle_sim_array['flag']==2)])} (decayed), n3={len(particle_sim_array[(particle_sim_array['flag']==3)])} (wall hit), n4={len(particle_sim_array[(particle_sim_array['flag']==4)])} (neutralized)")
                # drift times
                drift_time_data = particle_sim_array["drift_time"][(particle_sim_array["flag"]==1)]
                n_bins = 100
                bin_range = [np.min(drift_time_data),np.max(drift_time_data)]
                cts, bin_edges, patches = plt.hist(drift_time_data, bins=n_bins, range=bin_range)
                cts = [int(c) for c in cts]
                bin_width = bin_edges[1]-bin_edges[0]
                bin_centers = [be+0.5*bin_width for be in bin_edges[:-1]]
                #drift_time_mean, drift_time_lower, drift_time_upper = 5, 5, 5
                drift_time_mean, drift_time_lower, drift_time_upper = monxeana.get_asymmetric_intervals_around_center_val_for_interpolated_discrete_distribution(
                    distribution_bin_centers = bin_centers,
                    distribution_counts = cts,
                    #flag_verbose = True,
                    ctr_max = 300)
                print(f"\tdrift_time: mean={drift_time_mean}, lower={drift_time_lower}, upper={drift_time_upper}")
                # nuclei
                print(f"\tnucleus: po218={len(particle_sim_array[(particle_sim_array['nucleus']=='po218')])}, pb214={len(particle_sim_array[(particle_sim_array['nucleus']=='pb214')])}, bi214={len(particle_sim_array[(particle_sim_array['nucleus']=='bi214')])}, po214={len(particle_sim_array[(particle_sim_array['nucleus']=='po214')])}")
                # efficiencies
                n_sim_corr = n_sim -len(particle_sim_array[(particle_sim_array["flag"]==3)])
                n_sim_corr = n_sim
                n_po218_on_sensor = len(particle_sim_array[
                    (particle_sim_array["flag"]==1) 
                    & (particle_sim_array["nucleus"]=="po218")
                ])
                n_po214_on_sensor = len(particle_sim_array[
                    (particle_sim_array["flag"]==1)
                ])
                eff_coll_po218 = n_po218_on_sensor/n_sim_corr
                eff_coll_po214 = n_po214_on_sensor/n_sim_corr
                
                # appending the final retrieved values
                simulation_outcome_tuple_list.append((
                    e_field_sim,
                    u_coll_v,
                    eff_coll_po218,
                    eff_coll_po214,
                    drift_time_mean,
                    drift_time_lower,
                    drift_time_upper,
                    neutral_halflife,
                ))
                print(f"\tcollection efficiencies: eff_po218={eff_coll_po218:.2f}, eff_po214={eff_coll_po214:.2f}\n")


    # writing efficiency simulation outcome to disk
    simulation_outcome_array = np.array(simulation_outcome_tuple_list, simulation_outcome_dtype)
    np.save(abspath_this_study +"simulation_outcome_array.npy", simulation_outcome_array)
    t_f = datetime.datetime.now()
    print(f"Running this simulation took {t_f-t_i} hours.")


# printing the simulation results
simulation_outcome_array = np.load(abspath_this_study +"simulation_outcome_array.npy")
for entry in simulation_outcome_array:
    print(f"{entry[0]}, {entry[1]}, {entry[7]}: po218={100*entry[2]:.2f}, po214={100*entry[3]:.2f}, time={entry[4]:.2f}-{entry[5]:.2f}+{entry[6]:.2f}")
```

```python

# This cell is used to nicely print the simulation results


simulation_outcome_array = np.load(abspath_this_study +"simulation_outcome_array.npy")
e_field_sim_list = list(set(list(simulation_outcome_array["e_field_sim"])))
u_coll_v_list = list(set(list(simulation_outcome_array["u_coll_v"])))
u_coll_v_list.sort()
neutral_halflife_list = list(set(list(simulation_outcome_array["neutral_halflife"])))

for e_field_sim in e_field_sim_list:
    print(f"\n\n\n\n######### {e_field_sim}:")
    simulation_outcome_array_e_field_sim = simulation_outcome_array[(simulation_outcome_array["e_field_sim"]==e_field_sim)]
    for neutral_halflife in neutral_halflife_list:
        print(f"\n\tneutral_halflife={neutral_halflife}")
        print(f"\t\tu_coll_v:".ljust(15) +"   ".join([f"{u_coll_v:.2f}".rjust(26) for u_coll_v in u_coll_v_list]))
        print("\t\t" +90*"-")
        for feature in ["eff_po218", "eff_po214", "drift_time"]:
            if feature in ["eff_po218", "eff_po214"]:
                feature_list = [simulation_outcome_array[(simulation_outcome_array["e_field_sim"]==e_field_sim) & (simulation_outcome_array["u_coll_v"]==u_coll_v) & (simulation_outcome_array["neutral_halflife"]==neutral_halflife)][feature][0] for u_coll_v in u_coll_v_list]
                print(f"\t\t{feature}:".ljust(16) +"  & ".join([f"{100*ftr:.2f}".rjust(25) for ftr in feature_list]))
            elif feature == "drift_time":
                drift_time_mean_list = [simulation_outcome_array[(simulation_outcome_array["e_field_sim"]==e_field_sim) & (simulation_outcome_array["u_coll_v"]==u_coll_v) & (simulation_outcome_array["neutral_halflife"]==neutral_halflife)]["drift_time_mean"][0] for u_coll_v in u_coll_v_list]
                drift_time_lower_list = [simulation_outcome_array[(simulation_outcome_array["e_field_sim"]==e_field_sim) & (simulation_outcome_array["u_coll_v"]==u_coll_v) & (simulation_outcome_array["neutral_halflife"]==neutral_halflife)]["drift_time_lower"][0] for u_coll_v in u_coll_v_list]
                drift_time_upper_list = [simulation_outcome_array[(simulation_outcome_array["e_field_sim"]==e_field_sim) & (simulation_outcome_array["u_coll_v"]==u_coll_v) & (simulation_outcome_array["neutral_halflife"]==neutral_halflife)]["drift_time_upper"][0] for u_coll_v in u_coll_v_list]
                print(f"\t\t{feature}:".ljust(16) +"  & ".join([
                    r"$" +
                    f"{drift_time_mean_list[i]:.2f}".rjust(5) +
                    r"_{-" +
                    f"{drift_time_lower_list[i]:.2f}".rjust(5) +
                    r"}^{+" +
                    f"{drift_time_upper_list[i]:.2f}".rjust(5) +
                    r"}$" for i in range(len(drift_time_mean_list))]))
```
